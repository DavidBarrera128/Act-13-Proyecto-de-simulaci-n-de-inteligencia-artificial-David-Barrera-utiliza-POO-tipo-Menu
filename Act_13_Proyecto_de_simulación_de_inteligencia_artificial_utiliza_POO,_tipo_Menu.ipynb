# Calse : Cheque

```python
class Cheque:
    def __init__(self):
        # 5 Atributos inicializados por defecto
        self.numero_cheque = 0
        self.beneficiario = "Sin asignar"
        self.monto = 0.0
        self.fecha_emision = "Sin fecha"
        self.estado = "No Emitido"  # Estado inicial predeterminado

    # Función 1: Para validar que el monto ingresado sea correcto
    def verificar_monto(self, monto):
        if monto > 0:
            return True
        else:
            return False

    # Función 2: Llena los datos del cheque y lo activa
    def emitir_cheque(self, numero, beneficiario, monto, fecha):
        if self.estado != "No Emitido":
            print("\n[ERROR] Este cheque ya ha sido procesado anteriormente y no se puede modificar.")
            return

        if self.verificar_monto(monto):
            self.numero_cheque = numero
            self.beneficiario = beneficiario
            self.monto = monto
            self.fecha_emision = fecha
            self.estado = "Pendiente"
            print(f"\n[ÉXITO] Cheque No. {self.numero_cheque} emitido correctamente a favor de {self.beneficiario}.")
        else:
            print("\n[ERROR] El monto debe ser una cantidad mayor a 0.")

    # Función 3: Realiza el cobro cambiando su estado
    def cobrar_cheque(self):
        if self.estado == "Pendiente":
            self.estado = "Cobrado"
            print(f"\n[ÉXITO] El cheque No. {self.numero_cheque} por ${self.monto:.2f} ha sido cobrado en ventanilla.")
        elif self.estado == "Cobrado":
            print("\n[ALERTA] Este cheque ya fue cobrado previamente.")
        elif self.estado == "Cancelado":
            print("\n[ERROR] No se puede cobrar un cheque que ha sido cancelado.")
        else:
            print("\n[ERROR] El cheque aún no ha sido emitido, no hay nada que cobrar.")

    # Función 4: Anula el cheque de manera permanente
    def cancelar_cheque(self):
        if self.estado == "Pendiente":
            self.estado = "Cancelado"
            print(f"\n[ÉXITO] El cheque No. {self.numero_cheque} ha sido cancelado correctamente.")
        elif self.estado == "Cobrado":
            print("\n[ERROR] No se puede cancelar un cheque que ya fue cobrado con anterioridad.")
        elif self.estado == "Cancelado":
            print("\n[ALERTA] El cheque ya se encuentra cancelado.")
        else:
            print("\n[ERROR] No se puede cancelar un cheque que no ha sido emitido.")

    # Función 5: Muestra en pantalla la información formateada del objeto
    def mostrar_informacion(self):
        print("\n=======================================")
        print("         DETALLES DEL CHEQUE           ")
        print("=======================================")
        print(f" Número de Cheque : {self.numero_cheque}")
        print(f" Beneficiario     : {self.beneficiario}")
        print(f" Monto            : ${self.monto:.2f}")
        print(f" Fecha de Emisión : {self.fecha_emision}")
        print(f" Estado Actual    : {self.estado}")
        print("=======================================")


# --- Bloque de menú interactivo (Simulador en Consola) ---
if __name__ == "__main__":
    # Creamos la instancia/objeto de la clase Cheque
    mi_cheque = Cheque()

    while True:
        print("\n=== SIMULADOR DE CHEQUES BANCARIOS ===")
        print("1. Emitir nuevo cheque")
        print("2. Cobrar cheque")
        print("3. Cancelar cheque")
        print("4. Mostrar información del cheque")
        print("5. Salir")

        opcion = input("Selecciona una opción (1-5): ")

        if opcion == "1":
            try:
                num = int(input("Introduce el número de cheque: "))
                nom = input("Introduce el nombre del beneficiario: ")
                cant = float(input("Introduce el monto del cheque: $"))
                fec = input("Introduce la fecha de emisión (DD/MM/AAAA): ")
                mi_cheque.emitir_cheque(num, nom, cant, fec)
            except ValueError:
                print("\n[ERROR] Por favor, ingresa datos numéricos válidos en el número y el monto.")

        elif opcion == "2":
            mi_cheque.cobrar_cheque()

        elif opcion == "3":
            mi_cheque.cancelar_cheque()

        elif opcion == "4":
            mi_cheque.mostrar_informacion()

        elif opcion == "5":
            print("\nGracias por utilizar el simulador bancario. ¡Hasta pronto!")
            break
        else:
            print("\n[ERROR] Opción no válida. Intenta de nuevo.")
    print("Programa Realizado por David Barrera 1388")
```
